```python
sns.scatterplot(x='sepal_length', y='sepal_width',
                hue='species', data=df, )
```

Puedo ver que la especie Iris Setosa tiene los sépalos más pequeños de largo pero más anchos.
Sin embargo, Iris Versicolor tiene el tamaño medio entre las otras dos especies.
Iris virginica, tiene los sépalos largos pero más pequeños que iris Setosa.
Identifico valores que se salen de la norma (outliers):
- Iris setosa veo que una de las flores estudiadas de esta especie su sépalo es mucho más pequeño de ancho.
- Iris versicolor una de las flores estudiadas tiene el sépalo mas estrecho de todas las flores estudiadas de cualquiera de las tres especies.
- Iris virginica dos flores estudiadas se salen de la media de ancho de sépalo.
Pienso que estos outliers pueden deberse a que se han estudiado flores en proceso de cremimiento distinto.

```python
sns.scatterplot(x='petal_length', y='petal_width',
                hue='species', data=df, )
```

En esta gráfica se puede ver mejor la diferencia de las especies en el marco del tamaño de los pétalos.

La especie de Iris setosa tiene los pétalos más pequeños.
La especie Iris Versicolor tiene el tamaño medio de los pétalos entre las dos especies.
La especie Iris Virginica tiene los pétalos más grandes de ancho y largo.
Puedo observar gran diferencia de Iris Setosa, que es la especie que más igual tiene los pétalos, respecto a a Iris Versicolor.

Lo que me lleva a la conclusión que se han guiado más en clasificar la especie (Setosa, Versicolor y Virginica) por el tamaño de largo y ancho de los pétalos, puesto que se ve mas las diferencias que en el gráfico de los sépalos.

Mapa de calor:
Vemos las correlaciones entre las variables, las más interesantes que me parecen son comparar el largo del pétalo con el ancho, después es interesante estudiar la correlación entre el largo del pétalo con el largo del sépalo, pero no veo correlación entre el ancho del pétalo y el ancho del sépalo.


```python
# Voy a cortar el mapa de calor
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(df.corr(), mask=mask, vmax=1., vmin=-1., center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
```
